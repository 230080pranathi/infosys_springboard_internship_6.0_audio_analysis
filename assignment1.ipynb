```python
from abc import ABC,abstractmethod

class User:
    def __init__(self,name,balance):
        self.name=name
        self.__balance=balance  #Private 

    def get_balance(self):
        return self.__balance
    
    def deduct_balance(self,amount):
        if amount<=self.__balance:
            self.__balance-=amount
            return True
        else:
            return False
        



class Payment(ABC):
    @abstractmethod
    def pay(self,amount):
        pass




class CardPayment(Payment):
    def __init__(self,user):
        self.user=user

    def pay(self,amount):
        if self.user.deduct_balance(amount):
            return f"Paid ₹{amount} using CARD. Remaining balance: ₹{self.user.get_balance()}"
        else:
            return "Card Payment Failed:Insufficient Balance"
        



        
class UPIPayment(Payment):
    def __init__(self,user):
        self.user=user

    def pay(self,amount):
        if self.user.deduct_balance(amount):
            return f"Paid ₹{amount} using UPI.Remaining balance: ₹{self.user.get_balance()} "
        else:
            return "UPI Payment Failed:Insufficient Balance"
        



user1=User("John",5000)

payments=[CardPayment(user1),UPIPayment(user1)]

print("-----Demonstrating Abstraction,Polymorphism and Encapsulation ----")

for p in payments:
    print(p.pay(1500))

```
